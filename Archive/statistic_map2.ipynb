```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import re
```

```python
rewards = []
repeats = [i+1 for i in range(3500)]
with open('QLearning_map2_rewards.dat') as txtfile:
    for line in txtfile:
        line = line.strip()
        rewards += [float(line)]
        
print(max(rewards))
print(min(rewards))
fig = plt.figure(figsize=(8,4))
fig.suptitle('map 2')
plt.plot(repeats, rewards)
plt.axis([0, 3500,-220.0, -20])
plt.savefig(re.sub(r'\.dat$', '.png','QLearning_map2_rewards.dat' ))
plt.show()

maxindex = np.argmax(rewards) + 1
print("first index to reach max: ", maxindex)
print("max reward: ", rewards[maxindex-1])
```

```python
filename = "QLearning_map2_stats.dat"
df = pd.read_csv(filename, sep='\s+', header=None, skiprows=0)

# Grab left-most column
ys = list(df.to_dict()[1].values())

# Create x-values 0-N for iteration count
# Plot values
plt.plot(df[0], ys, 'b')
plt.xlabel('Iterations')
plt.ylabel('Number of Errors')
plt.title('Q-Learning')

# Plot is saved as PNG with same name as DAT file
plt.savefig(re.sub(r'\.dat$', '.png', filename))
plt.show()
```
