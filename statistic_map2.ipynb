```python
import numpy as np
import matplotlib.pyplot as plt

rewards = []
repeats = [i+1 for i in range(3500)]
with open('QLearning_map2_rewards.dat') as txtfile:
    for line in txtfile:
        line = line.strip()
        rewards += [float(line)]
        
print(max(rewards))
print(min(rewards))
fig = plt.figure(figsize=(8,4))
fig.suptitle('map 2')
plt.plot(repeats, rewards)
plt.axis([0, 3500,-250.0, -20])
plt.show()

maxindex = np.argmax(rewards) + 1
print("first index to reach max: ", maxindex)
print("max reward: ", rewards[maxindex-1])
```

```python
np.random.seed(0)
%matplotlib inline

DQN_errorlog = np.loadtxt('QLearning_map2_stats.dat') #(X,Y)
DQN_errorX = []
DQN_errorY = []

for i in range(len(DQN_errorlog)):
    DQN_errorX.append(DQN_errorlog[i][0])
    DQN_errorY.append(DQN_errorlog[i][1])
    
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
ax.scatter(DQN_errorX, DQN_errorY, s=100, marker='.', color='red', alpha=0.75, label='Number of Error')

#calc trend line
z = np.polyfit(DQN_errorX, np.sqrt(DQN_errorY), 2)
p = np.poly1d(z)
ax.plot(DQN_errorX, p(DQN_errorX), lw=3, color='blue', alpha=0.75, label='Trend')
```
