```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import re
import scipy as sp
from scipy.optimize import curve_fit
np.random.seed(0)
%matplotlib inline

def exponenial_func(x, a, b, c):
    return a*np.exp(-b*x)+c

y = np.loadtxt('DeepQLearning2Stats/DeepQLearning2_map1_rewards.dat') #(X,Y)
x = []

for i in range(len(y)):
    x.append(i)
    
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax.bar(x, y)

plt.axis([-50, 2510, -25, -75])
plt.xlabel('Number of Episodes')
plt.ylabel('Reward Value (negative values)')
plt.title('Map1 Reward per Episode')
plt.plot()
```
