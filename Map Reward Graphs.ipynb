```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import re
import scipy as sp
from scipy.optimize import curve_fit
np.random.seed(0)
%matplotlib inline

def exponenial_func(x, a, b, c):
    return a*np.exp(-b*x)+c

DQN_errorlog = np.loadtxt('DeepQLearning2Stats/DeepQLearning2_map1_stats.dat') #(X,Y)
x = []
y = []

for i in range(len(DQN_errorlog)):
    x.append(DQN_errorlog[i][0])
    y.append(DQN_errorlog[i][1])

popt, pcov = curve_fit(exponenial_func, x, y, p0=(1, 1e-6, 1))

xx = np.linspace(0, 3000, 1000)
yy = exponenial_func(xx, *popt)

fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax.scatter(x, y, s=80, marker='.', color='red', alpha=0.75, label='Number of Error')

ax.plot(xx, yy, lw=3, color='blue', alpha=0.75, label='Prediction')

plt.axis([-50, 2510, -0.5, 40])
plt.xlabel('Number of Episodes')
plt.ylabel('Number of Errors')
plt.title('Map1 Error-Rate')
plt.plot()
```
